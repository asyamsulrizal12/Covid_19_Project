## 📌 COVID-19 in Indonesia: Data Insights

**Major Waves**
  - Two significant peaks observed: mid-2021 (Delta variant) and early–mid 2022 (Omicron variant).
  - Daily cases reached extreme highs (~60k), followed by sharp declines.

**Long-Term Trend**
  - After 2022, daily cases steadily decreased, indicating a more controlled phase.

**Weekly Seasonality**
  - Clear oscillating weekly pattern, likely due to reporting cycles (weekend effect).

**Distribution & Outliers**
  - Highly skewed data: most days with low case counts, but extreme spikes during waves.
  - Outliers highlight sudden surges and stress the need for robust modeling.

**Correlation**
  - Strong positive correlation between cumulative cases and deaths (~0.98).
  - Negative correlation between cases and recoveries, reflecting reporting delays.

**Feature Engineering Impact**
  - **Lag & Rolling Features**: capture temporal dependencies and smooth trends.
  - **Growth Rate**: reveals acceleration/deceleration in spread dynamics.
  - **Log Transformation**: reduces skewness, stabilizes data for modeling.

---

### Key Takeaways
- Indonesia experienced **two major COVID-19 waves** with clear weekly seasonality.
- Data is skewed with extreme outliers → preprocessing is essential.
- Strong link between cases and deaths → lagged features can improve predictions.
- Feature engineering prepares the dataset for robust time-series models (ARIMA, Prophet, LSTM).
